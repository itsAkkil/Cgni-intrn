```python
y = df['Aggregate rating']
X = df.drop(['Aggregate rating','Restaurant Name','Address', 'City','City', 'Address',
       'Locality', 'Locality Verbose', 'Cuisines', 'Currency'],axis=1)
```

```python
# ## OLS Linear Regression
import statsmodels.api as sm
```

```python
print(sm.OLS(y,ol).fit().summary())
```

```python
## Lets do some viualization
import seaborn as sns
sns.pairplot(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2529)
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

import statsmodels.api as sm
```

```python
categorical_cols = X.select_dtypes(include=['object','bool']).columns
```

```python
numerical_cols = X.select_dtypes(include=['int64','float64']).columns
```

```python
ohe = OneHotEncoder(drop='first')
```

```python
ss = StandardScaler()
```

```python
preprocessor=ColumnTransformer(transformers=[("num",ss,numerical_cols), ("cat",ohe,categorical_cols),])
```

```python
model = Pipeline(steps=[("data_prep", preprocessor ), ("Linear_class",LinearRegression() )])
```

```python
model.fit(X_train, y_train)
```

```python
lr_sml=model.fit(X_train, y_train)
```

```python
lr_sml.score(X_test,y_test)
```

```python
y_pred= model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error
```

```python
## Performance Metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
#display adjusted R-squared
print(1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
